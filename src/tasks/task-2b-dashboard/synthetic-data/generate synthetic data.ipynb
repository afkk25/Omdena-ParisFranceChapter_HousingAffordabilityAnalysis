```python
import random
import pandas as pd
```

```python
def generate_address():
    # Function to generate a random address
    streets = ['Rue de Rivoli', 'Avenue des Champs-Élysées', 'Boulevard Saint-Germain', 'Rue Saint-Honoré', 'Rue de la Paix', 
               'Avenue Montaigne', 'Rue de la République', 'Boulevard Haussmann', 'Rue de Vaugirard', 'Rue du Faubourg Saint-Honoré']
    return f'{random.randint(1, 200)} {random.choice(streets)}'

def generate_district():
    # Function to generate a random district/arrondissement
    return random.randint(1, 20)

def generate_zipcode():
    # Function to generate a random zipcode
    return f'75{random.randint(1, 20):02d}'

def generate_rent():
    # Function to generate a random rent
    return random.randint(1000, 10000)

def generate_cost():
    # Function to generate a random cost of property
    return random.randint(100000, 5000000)

def generate_maintenance():
    # Function to generate a random maintenance cost
    # 10% of the rent of the cost of property
    return random.randint(30, 200)

def generate_rooms():
    # Function to generate a random number of rooms
    return random.randint(1, 5)

def generate_size():
    # Function to generate a random size in square feet
    return random.randint(300, 2000)

def generate_property_type():
    # Function to generate a random property type
    return random.choice(['Apartment', 'Duplex', 'Villa'])

def co_living():
    # Returns wether Co-living is allowed or not
    return random.randint(0 , 1)

def are_pets_allowed():
    # Returns wether pets are  allowed or not
    return random.randint(0 , 1)

def lease_duration():
    # Returns the lease duration in months
    return random.randint(1 , 36)

# https://www.w3schools.com/python/ref_random_uniform.asp
# Paris co-ordinates are googled
def generate_latitude():
    return round(random.uniform(48.80, 48.90), 6)

def generate_longitude():
    return round(random.uniform(2.25, 2.45), 6)
```

```python
# Generate data
data = {
    'Address': [generate_address() for _ in range(500)],
    'District/Arrondissement': [generate_district() for _ in range(500)],
    'Zipcode': [generate_zipcode() for _ in range(500)],
    'Rent': [generate_rent() for _ in range(500)],
    'Cost': [generate_cost() for _ in range(500)],
    'Maintenance': [generate_maintenance() for _ in range(500)],
    'Number of rooms': [generate_rooms() for _ in range(500)],
    'Size': [generate_size() for _ in range(500)],
    'Property Type': [generate_property_type() for _ in range(500)],
    'Co-Living': [co_living() for _ in range(500)],
    'Pets allowed': [are_pets_allowed() for _ in range(500)],
    'Lease Duration': [lease_duration() for _ in range(500)],
    'Latitude': [generate_latitude() for _ in range(500)],
    'Longitude': [generate_longitude() for _ in range(500)]
}

# Create DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
preference_data = {
    'Address': [generate_address() for _ in range(100)],
    'District/Arrondissement': [generate_district() for _ in range(100)],
    'Number of rooms': [generate_rooms() for _ in range(100)],
    'Size': [generate_size() for _ in range(100)],
    'Property Type': [generate_property_type() for _ in range(100)],
    'Lease_duration': [lease_duration() for _ in range(100)],
    'Co-Living': [co_living() for _ in range(100)],
    'Rent': [generate_rent() for _ in range(100)],
    'Pets allowed': [are_pets_allowed() for _ in range(100)]
}

pf_df = pd.DataFrame(preference_data)
pf_df.head()
```
